# Generadores

Funciones que producen una secuencia de valores bajo demanda (lazy evaluation) sin crear listas completas en memoria.

## Resumen - ¿Por qué Generadores?

### Conceptos Clave
- **Lazy evaluation:** Valores generados bajo demanda (cuando se necesitan)
- **Bajo uso de memoria:** No almacena toda la secuencia en memoria
- **Sintaxis simple:** Usa `yield` en lugar de `return`
- **Iterables:** Funcionan con `for`, `next()`, list comprehensions
- **Eficiencia:** Ideal para datos grandes o infinitos

## 1️⃣ Concepto Básico de Generadores

Un **generador** es una función que produce valores uno a uno, bajo demanda.

### Diferencia función normal vs generador:

```python
# Función normal: crea lista en memoria
def numeros_normales(n):
    resultado = []
    for i in range(n):
        resultado.append(i)
    return resultado

# Generador: produce valores bajo demanda
def numeros_generador(n):
    for i in range(n):
        yield i
```

**Ventaja:** Si n=1,000,000, el generador usa casi nada de memoria, mientras que la lista necesita GB.

````python
import sys

# Comparación: función normal vs generador

def numeros_list(n):
    resultado = []
    for i in range(n):
        resultado.append(i)
    return resultado

def numeros_gen(n):
    for i in range(n):
        yield i

lista = numeros_list(100)
gen = numeros_gen(100)

print(f"Tamaño lista:      {sys.getsizeof(lista)} bytes")
print(f"Tamaño generador:  {sys.getsizeof(gen)} bytes")
print()
print(f"Tipo lista:       {type(lista)}")
print(f"Tipo generador:   {type(gen)}")
print()
print("Valores generador (primeros 5):")
for i, valor in enumerate(gen):
    if i >= 5:
        break
    print(f"  {valor}", end="")
print("  ...")
````

## 2️⃣ La Palabra Clave `yield`

**`yield`** es como `return`, pero:
- Devuelve un valor
- **Pausa** la función (guarda el estado)
- **Reanuda** cuando se pide el siguiente valor

### Flujo de ejecución con yield:

```python
def contador(n):
    print("Start")
    for i in range(n):
        print(f"Antes de yield {i}")
        yield i
        print(f"Después de yield {i}")

gen = contador(3)
print(next(gen))  # "Start", "Antes de yield 0" → 0
print(next(gen))  # "Después de yield 0", "Antes de yield 1" → 1
print(next(gen))  # "Después de yield 1", "Antes de yield 2" → 2
```

````python
# Visualizar el flujo de yield

def contador(n):
    print("  Inicializando generador")
    for i in range(n):
        print(f"  Antes de yield {i}")
        yield i
        print(f"  Después de yield {i}")
    print("  Generador finalizado")

print("Creando generador...")
gen = contador(3)
print()

print("Llamada 1 a next():")
print(f"  Valor: {next(gen)}")
print()

print("Llamada 2 a next():")
print(f"  Valor: {next(gen)}")
print()

print("Llamada 3 a next():")
print(f"  Valor: {next(gen)}")
print()

print("Llamada 4 a next() (agotado):")
try:
    next(gen)
except StopIteration:
    print("  StopIteration - No hay más valores")
````

## 3️⃣ Generator Expressions

Sintaxis compacta similar a list comprehensions, pero sin crear lista completa.

### Comparación:

```python
# List comprehension: crea lista completa
lista = [x**2 for x in range(5)]  # [0, 1, 4, 9, 16]

# Generator expression: genera bajo demanda
gen = (x**2 for x in range(5))  # <generator object>

# Ambas producen los mismos valores, pero el generador usa menos memoria
```

````python
import sys

# List comprehension
lista_cuadrados = [x**2 for x in range(1000)]
print(f"List comprehension - Tamaño: {sys.getsizeof(lista_cuadrados)} bytes")
print(f"  Primeros 5 elementos: {lista_cuadrados[:5]}")
print()

# Generator expression
gen_cuadrados = (x**2 for x in range(1000))
print(f"Generator expression - Tamaño: {sys.getsizeof(gen_cuadrados)} bytes")
print(f"  Primeros 5 elementos: {list(gen_cuadrados)[:5]}")
print()

# Generator expression con filter
print("Generator expression con filter (pares):")
pares = (x for x in range(20) if x % 2 == 0)
print(f"  {list(pares)}")
````

## 4️⃣ `range()`, `enumerate()` y `zip()` - Generadores Incorporados

Funciones que producen generadores o iterables.

### `range(start, stop, step)`
Generador de números enteros.

### `enumerate(iterable, start=0)`
Generador de tuplas (índice, valor).

### `zip(*iterables)`
Generador que combina múltiples iterables.

````python
# range(), enumerate(), zip()

print("1️⃣ range():")
for num in range(0, 10, 2):
    print(f"  {num}", end="")
print("\n")

print("2️⃣ enumerate():")
frutas = ['manzana', 'plátano', 'cereza']
for indice, fruta in enumerate(frutas):
    print(f"  [{indice}] {fruta}")
print()

print("3️⃣ zip():")
nombres = ['Juan', 'María', 'Pedro']
edades = [25, 30, 35]
ciudades = ['Madrid', 'Barcelona', 'Valencia']

for nombre, edad, ciudad in zip(nombres, edades, ciudades):
    print(f"  {nombre}: {edad} años en {ciudad}")
````

## 5️⃣ Composición de Generadores

Encadenar múltiples generadores para procesar datos eficientemente.

### Patrón:

```python
def gen1():
    for i in range(5):
        yield i

def gen2(gen):
    for valor in gen:
        yield valor * 2

# Composición: gen2(gen1()) procesa valores de gen1()
resultado = gen2(gen1())
```

**Ventaja:** Los datos se procesan bajo demanda, sin crear listas intermedias.

````python
# Composición de generadores

def rango_generador(start, end):
    print(f"  Generando rango ({start}...{end})")
    for i in range(start, end):
        print(f"    Yielding {i}")
        yield i

def duplicar(gen):
    print(f"  Procesando (duplicar)")
    for valor in gen:
        print(f"    Duplicando {valor} -> {valor*2}")
        yield valor * 2

def filtrar_pares(gen):
    print(f"  Procesando (filtrar pares)")
    for valor in gen:
        if valor % 2 == 0:
            print(f"    Aceptando {valor}")
            yield valor
        else:
            print(f"    Rechazando {valor}")

print("Composición: rango -> duplicar -> filtrar pares\n")
resultado = filtrar_pares(duplicar(rango_generador(1, 5)))

print("\nConsumir generador:")
print(f"Valores finales: {list(resultado)}")
````

## 6️⃣ Performance: Generadores vs Listas

### Benchmark: Procesar datos grandes

```python
# Lista: carga TODO en memoria
lista = [x**2 for x in range(1000000)]
resultado = [x for x in lista if x % 2 == 0]

# Generador: procesa bajo demanda
gen = (x**2 for x in range(1000000))
resultado = (x for x in gen if x % 2 == 0)
```

**Conclusión:** Generadores son ideales cuando:
- ✅ Procesas datos muy grandes
- ✅ Solo necesitas los primeros N valores
- ✅ Tienes secuencias infinitas
- ✅ Quieres código lazy y eficiente

````python
import time
import sys

print("Benchmark: Lista vs Generador\n")

n = 1000000

# LISTA
print("1️⃣ Creando lista con 1,000,000 cuadrados...")
start = time.time()
lista = [x**2 for x in range(n)]
tiempo_lista = time.time() - start
tamaño_lista = sys.getsizeof(lista)
print(f"   Tiempo: {tiempo_lista:.4f}s")
print(f"   Tamaño: {tamaño_lista / (1024**2):.2f} MB\n")

# GENERADOR
print("2️⃣ Creando generador con 1,000,000 cuadrados...")
start = time.time()
gen = (x**2 for x in range(n))
tiempo_gen = time.time() - start
tamaño_gen = sys.getsizeof(gen)
print(f"   Tiempo: {tiempo_gen:.6f}s")
print(f"   Tamaño: {tamaño_gen} bytes\n")

print("Resultados:")
print(f"   Lista es {tamaño_lista / tamaño_gen:,.0f}x más grande")
print(f"   Generador es {tiempo_lista / tiempo_gen:,.0f}x más rápido de crear")
print()
print("Conclusión: Generadores son MUCHO más eficientes para datos grandes")
````
